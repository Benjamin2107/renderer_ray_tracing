```python
from src.utilities import Vec3, Ray, ray_color, Camera, color
from src.objects import Sphere, HittableList
```

```python
camera = Camera()

max_color = 255
image_width = 400
image_height = int(camera.get__height_from_width(image_width))

viewport_height = 2.0
viewport_width = camera.get_width_from_height(viewport_height)

horizontal = Vec3(viewport_width, 0, 0)
vertical = Vec3(0, viewport_height, 0)
lower_left_corner = camera.origin - horizontal / 2 - vertical / 2 - Vec3(0, 0, camera.focal_length)
```

```python
#red
sphere1 = Sphere(Vec3(0, -0.3, -1), 0.2, Vec3(1, 0, 0))
#green
sphere2 = Sphere(Vec3(0, -100.5, -1), 100, Vec3(0, 1, 0))
#blue
sphere3 = Sphere(Vec3(0.5, -0.1, -0.75), 0.2, Vec3(0, 0, 1))
#turquoise
sphere4 = Sphere(Vec3(-0.5, -0.1, -0.75), 0.3, Vec3(0, 1, 1))
#yellow
sphere5 = Sphere(Vec3(0, 0.5, -2), 1, Vec3(1, 1, 0))

hittables = HittableList([sphere1, sphere2, sphere3, sphere4, sphere5])
```

```python
"""
Rendering Loop, adapted to needs for this task
"""
with open("3_output_image/img_five_spheres.ppm", "w") as img:
    img.write(f"P3\n{image_width} {image_height}\n{max_color}\n")

    for j in range(image_height - 1, -1, -1):
        print(f"Lines left:{j}")
        for i in range(image_width):
            u = float(i) / (image_width-1)
            v = float(j) / (image_height-1)
            ray = Ray(camera.origin, lower_left_corner + horizontal*u + vertical*v -camera.origin)
            pixel_color = color(ray_color(ray, hittables), 255)
            img.write(f"{int(pixel_color.get_x())} ")
            img.write(f"{int(pixel_color.get_y())} ")
            img.write(f"{int(pixel_color.get_z())} \n")

    img.close()
```
